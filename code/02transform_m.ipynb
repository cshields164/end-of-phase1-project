```python
# Since I am only analyzing parental status and alcohol affecting student performance,
# I can drop all columns except alcohol and parental status as well as the grading periods and failures.

drop_list = ['failures', 'G1', 'G2', 'G3', 'Pstatus', 'Dalc', 'Walc']
df_new_m = df_m.drop(df_m.columns.difference(drop_list), axis = 1)
df_new_m
```

```python
df_new_m.shape
```

```python
# We can also test to see how the G1, G2, and G3 columns
# would look if they were combined together.

sum_grades_m = df_new_m['G1'] + df_new_m['G2'] + df_new_m['G3']
print("Sum of grades for Portuguese class students:", sum_grades_m)
```

```python
# the newly transformed dataset will be saved in this file name

df_new_m.to_csv('../data/student-mat-cleaned.csv')
```
