```python
import pandas as pd
import numpy as np
```

```python
# loading our dataframes

df_p = pd.read_csv('../data/student-por.csv', delimiter=';')
```

```python
# Since I am only analyzing parental status and alcohol affecting student performance,
# I can drop all columns except alcohol and parental status as well as the grading periods and failures.

drop_list = ['failures', 'G1', 'G2', 'G3', 'Pstatus', 'Dalc', 'Walc']
df_new_p = df_p.drop(df_p.columns.difference(drop_list), axis = 1)
df_new_p
```

```python
df_new_p.shape
```

```python
# We can also test to see how the G1, G2, and G3 columns
# would look if they were combined together.

sum_grades_p = df_new_p['G1'] + df_new_p['G2'] + df_new_p['G3']
print("Sum of grades for Portuguese class students:", sum_grades_p)
```

```python
# add this column into our dataframe
df_new_p['sum_grades_p'] = ['sum_grades_p']
df_new_p.head()
```

```python
# the newly transformed dataset will be saved in this file name

df_new_p.to_csv('../data/student-por-cleaned.csv')
```
